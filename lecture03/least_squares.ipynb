```python
%matplotlib notebook
import pylab
pylab.rcParams['figure.figsize'] = (10, 10)

from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generate some fake one-dimensional data
np.random.seed(1)
N = 50
D = 1
x = np.random.randn(N, D)
X = np.c_[np.ones(N), x]
w = np.random.randn(D + 1)
sigma2 = 0.1
y = np.dot(X, w) + np.random.randn(N) * sigma2
```

```python
h = .01
xx, yy = np.meshgrid(np.arange(-5, 5, h), np.arange(-5, 5, h))
W = np.c_[xx.ravel(), yy.ravel()]
X = np.c_[np.ones(N), x]
yhats = np.dot(X, W.T)
ress = yhats - y.reshape(N, 1)
losses = (.5/N) * np.linalg.norm(ress, axis=0)**2
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('$w_0$', fontsize=20)
ax.set_ylabel('$w_1$', fontsize=20)
ax.set_zlabel('Loss', fontsize=20)
ax.plot_surface(xx, yy, losses.reshape(xx.shape), cmap=cm.coolwarm)
cset = ax.contour(xx, yy, losses.reshape(xx.shape), zdir='z', offset=0, cmap=cm.coolwarm)
plt.show()
```
